```python
import pandas as pd
#Importacion de los datos del fichero indicado
df_datos = pd.read_csv(r"C:\Users\Christian\Downloads\Evaluación Estudiantil de la Calidad Educativa (SEEQ).csv")
```

```python
df_datos
```

```python
columnas_a_eliminar = ['Marca temporal', 'Por favor dinos cuales son tus Apellidos y Nombres']
df_datos = df_datos.drop(columns=columnas_a_eliminar)
```

```python
df_datos
```

```python
df_datos = df_datos.drop(df_datos.columns[0], axis=1)
```

```python
df_datos
```

```python
num_columnas = len(df_datos.columns)

```

```python
num_columnas
```

```python
# Renombrar las columnas desde la segunda en adelante
for i in range(1, num_columnas):
    nombre_nuevo = f"Q{i}"  # Construir el nombre de la columna
    df_datos = df_datos.rename(columns={df_datos.columns[i]: nombre_nuevo})
```

```python
df_datos
```

```python
df_datos = df_datos.rename(columns={df_datos.columns[0]: 'Filter'})
```

```python
df_datos
```

```python
df_ordenado = df_datos.sort_values(by=df_datos.columns[0])
```

```python
df_ordenado
```

```python
df_ordenado['Filter'] = df_ordenado['Filter'].replace({'Grupo con clase tradicional': 'Traditional Class',
                                                       'Grupo con dispositivo': 'Tech Device Class'})
```

```python
df_ordenado
```

```python
df_final = pd.DataFrame(df_ordenado)
```

```python
medias_por_filtro = df_final.groupby(df_final.columns[0]).mean()
```

```python
medias_por_filtro
```

```python
import matplotlib.pyplot as plt
```

```python
columnas_interes = medias_por_filtro.columns[0:4]

# Crear la gráfica de barras
plt.figure(figsize=(10, 6))
ax = plt.gca()  # Obtener el eje actual
ax.grid(axis='y') 
ax = medias_por_filtro[columnas_interes].plot(kind='bar', ax=plt.gca(), fontsize=20)  # Utilizar el mismo eje de la gráfica actual
plt.xlabel('', fontsize=20)
plt.ylabel('Mean Score "[about 5]"', fontsize=20,)
plt.legend(title='Questions', fontsize=14)
plt.xticks(rotation=0,fontsize=20)
plt.yticks(fontsize=20)

# Agregar etiquetas con los valores de cada barra
for p in ax.patches:
    ax.annotate(f"{p.get_height():.1f}", (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom', fontsize=18, color='black', xytext=(0, 10), textcoords='offset points')
plt.margins(y=0.1)
# Cambiar la fuente a Times New Roman
plt.rcParams['font.family'] = 'Times New Roman'
plt.tight_layout()  # Ajustar el diseño para evitar superposiciones
plt.show()
```
